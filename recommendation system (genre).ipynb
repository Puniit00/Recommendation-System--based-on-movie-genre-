```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

# Loading movies dataset

```python
movie = pd.read_csv('movies/movie.csv',sep=',',usecols = ['title','genres']);
```

```python
movie
```

```python
movie['genres'] = movie['genres'].str.split('|');
```

```python
movie['genres'] = movie['genres'].fillna("").astype('str')#converting genres to string
```

```python
movie['genres']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
#using tfidf vectorizer for removing stop words
tf = TfidfVectorizer(analyzer = 'word',ngram_range=(1,2),stop_words = 'english')
matrix = tf.fit_transform(movie['genres'])
matrix.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim = cosine_similarity(matrix,matrix)

cosine_sim[:5,:5]
```

```python
titles = movie['title'];
indices = pd.Series(movie.index,index = movie['title'])

def genre_recommendation(title):
    
        idx = indices[title]
        sim_score = list(enumerate(cosine_sim[idx]))
        sim_score = sorted(sim_score,key = lambda x:x[1],reverse = True)
        sim_score = sim_score[1:21]
        movies = [i[0] for i in sim_score]
        return titles.iloc[movies]
    
```

```python
genre_recommendation('Toy Story (1995)')
```

```python
df = pd.DataFrame(genre_recommendation('Toy Story (1995)'))
df.to_csv(path_or_buf="Solution_recsys.csv",index=False,encoding='utf-8')

```
